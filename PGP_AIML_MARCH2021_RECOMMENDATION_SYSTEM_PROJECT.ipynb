```python
algo = KNNWithMeans(k=100, sim_options={'name': 'cosine', 'user_based': True})
algo.fit(trainset)
```

```python
from surprise import SVD
from surprise import accuracy
```

```python
svd_model = SVD(n_factors=50)
svd_model.fit(trainset)
test_pred = svd_model.test(testset)
```
