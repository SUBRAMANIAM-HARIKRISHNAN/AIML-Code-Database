# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (5 points)

```python
df=pd.read_csv("C:/Users/HP/Downloads/Laliga.csv", header= 1,)
df.head(5)
df.rename({"T" : "Qualified"},axis = 1,inplace=True)
def data_cleaning(column):
    value=column
    if value == '-':
        return 0
    else:
        return value
    
for i in ('Pos', 'Team', 'Seasons', 'Points', 'GamesPlayed', 'GamesWon',
       'GamesDrawn', 'GamesLost', 'GoalsFor', 'GoalsAgainst', 'Champion',
       'Runner-up', 'Third', 'Fourth', 'Fifth', 'Sixth', 'Qualified', 'Debut',
       'Since/LastApp', 'BestPosition'):
       df[i]= df[i].apply(data_cleaning)
df =df.apply(pd.to_numeric,errors='ignore')
```

# 2. Print all the teams which have started playing between 1930-1980. Use “Debut” column. (Include both the years - 1930 and 1980) (5 points)

```python
new = df["Debut"].str.split("-", n = 1, expand = True)
df["Debut Season Start"]=new[0]
df["Debut Season End"]=new[1]
df.drop(["Debut Season End"],axis=1,inplace=True)
df["Debut Season Start"]=pd.to_datetime(df["Debut Season Start"],format="%Y")
TEAMS= (df['Debut Season Start'] > '1929') & (df['Debut Season Start'] < '1981')
df.loc[TEAMS]
```

# 3. Print the list of teams which came Top 5 in terms of points (2.5 points)

```python
Top5=df.nlargest(5, ['Points'])
print("The Top 5 Teams Based on Points are")
Top5[["Team","Points"]]
```

# 4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)
Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count(colo):
    g_For=colo[0]
    g_against=colo[1]
    diff=g_For-g_against
    return diff

df['Goal_diff_count']=df[['GoalsFor','GoalsAgainst']].apply(Goal_diff_count,axis=1)
Max_Goals=df['Goal_diff_count'].max()
Min_Goals=df['Goal_diff_count'].min()
Min_Team=df.iloc[df['Goal_diff_count'].idxmin(),1]
Max_Team=df.iloc[df['Goal_diff_count'].idxmax(),1]
print(Max_Team +" " + 'has the highest goal difference of', Max_Goals)
print(Min_Team +" " + 'has the lowest goal difference of', Min_Goals)
```


# 5. Create a new column with the name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which have the highest Winning percentage




```python
def Winning_Percentage(colo1):
    win=colo1[0]
    play=colo1[1]
    if ((win==0) or (play==0)):
        return 0
    else:
        percent=round(((win/play)*100),2)
        return percent

df['Winning Percentage']=df[['GamesWon','GamesPlayed']].apply(Winning_Percentage,axis=1)
Top5Win=df.nlargest(5, ['Winning Percentage'])
Top5Win[["Team","Winning Percentage"]]
```

# 6. Group teams based on their “Best position” and print the sum of their points for all positions (7.5 points)
